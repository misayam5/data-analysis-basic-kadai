```python
# Step1 matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する。
import matplotlib.pyplot as plt
import numpy as np

# Step2 6個のグラフを表示する領域を作成.
# Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述
fig, axes = plt.subplots(2, 3, tight_layout=True, figsize=(12, 8))

# Step3 各領域にグラフを描画する

# --- [0, 0] 折れ線グラフ ---
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title("折れ線グラフ")

# --- [0, 1] sin関数 ---
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)
axes[0, 1].set_title("sin関数")

# --- [0, 2] ヒストグラム ---
# 平均15, 標準偏差5の乱数を1000個生成
x2 = np.random.normal(15, 5, 1000)
axes[0, 2].hist(x2, bins=30)
axes[0, 2].set_title("ヒストグラム")

# --- [1, 0] 散布図 ---
# 100個のランダムな点プロット
x3 = np.random.rand(100)
y3 = np.random.rand(100)
axes[1, 0].scatter(x3, y3)
axes[1, 0].set_title("散布図")

# 残りの [1, 1] と [1, 2] は空欄でOK（課題の指示通り）

# Step4. グラフを表示
plt.show()
```
