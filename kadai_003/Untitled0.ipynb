```python
import numpy as np

# Step2 AとBの行列を格納する多次元配列のコードを記述する
a = np.array([[0,2,4],[1,3,5]])

# Bの形を「3行2列」にする（Aの列数3に合わせるため）
b = np.array([[0,1],[2,3],[4,5]])

# Step3 行列の積の計算をするコードを記述する
result = np.dot(a, b)

# Step4 計算結果を表示するコードを記述する
print("行列の積の結果")
print(result)

# Step5 行列の要素の最大値を求めるコードを記述する
print("最大値")
print(result.max())
```
